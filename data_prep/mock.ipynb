```python
toxic_ready = pd.read_csv(path+'all_toxic_reply_ready.csv')
cnt_ready = toxic_ready.groupby('user_id').count().reset_index()
cnt_ready.drop(cnt_ready[cnt_ready['tweet_id'] < 10].index, inplace=True)
print(cnt_ready.shape)
toxic_ready_repliers = set(cnt_ready['user_id'].tolist())
len(toxic_ready_repliers)
```

```python
nontoxic_ready = pd.read_csv(path+'all_nontoxic_reply_ready.csv')
print(toxic_ready.shape, nontoxic_ready.shape)
```

```python
nontoxic_sampled = nontoxic_ready.sample(frac=0.25)
```

```python
cnt_sampled = nontoxic_sampled.groupby('user_id').count().reset_index()
cnt_sampled.drop(cnt_sampled[cnt_sampled['tweet_id'] < 10].index, inplace=True)
print(cnt_sampled.shape)
nontoxic_sampled_repliers = set(cnt_sampled['user_id'].tolist())
len(nontoxic_sampled_repliers)
```

```python
combined = toxic_ready_repliers.union(nontoxic_sampled_repliers)
len(combined)
```

```python
sampled = pd.concat([toxic_ready, nontoxic_sampled], ignore_index=True)
cnt_ = sampled.groupby('user_id').count().reset_index()
cnt_.drop(cnt_[cnt_['tweet_id'] < 10].index, inplace=True)
print(cnt_.shape)
sampled_repliers = set(cnt_['user_id'].tolist())
len(sampled_repliers)
```

```python
all_r = list(sampled_repliers)
all_r
```

```python
dfs = []
for r in all_r:
    try:
        tox = toxic_ready[toxic_ready['user_id']==r].copy()
        tox['from'] = 'tox'
        dfs.append(tox)
    except:
        print(r, 'no toxic replies')
    try:
        nontox = nontoxic_sampled[nontoxic_ready['user_id']==r].copy()
        nontox['from'] = 'nontox'
        dfs.append(nontox)
    except:
        print(r, 'no nontoxic replies')

all_df = pd.concat(dfs, ignore_index=True)
all_df.head()
```
